```python
from collections import Counter
from functools import reduce

# Problem 1: Element appearing maximum number of times
def max_occurrence(arr):
    count = Counter(arr)
    max_elem = max(count.keys(), key=lambda x: count[x])
    return max_elem
print(max_occurrence([1, 3, 2, 2, 3, 3, 1]))  


# Problem 2: Missing number in the list
def find_missing_number(arr, n):
    total_sum = n * (n + 1) // 2
    arr_sum = sum(arr)
    missing_number = total_sum - arr_sum
    return missing_number
print(find_missing_number([1, 2, 4, 6, 3, 7, 8], 8))  


# Problem 3: Element occurring odd number of times
def find_odd_occurrence(arr):
    result = reduce(lambda x, y: x ^ y, arr)
    return result
print(find_odd_occurrence([1, 2, 3, 2, 3, 1, 3]))  # Problem 3


# Problem 4: Two elements whose sum is equal to a given element K
def find_two_sum(arr, K):
    seen = set()
    for number in arr:
        complement = K - number
        if complement in seen:
            return (number, complement)
        seen.add(number)
    return None
print(find_two_sum([2, 7, 11, 15], 9))  # Problem 4


# Problem 5: Two numbers whose sum is closest to 0
def closest_to_zero(arr):
    arr.sort()
    left, right = 0, len(arr) - 1
    min_sum = float('inf')
    result = (arr[left], arr[right])
    while left < right:
        curr_sum = arr[left] + arr[right]
        if abs(curr_sum) < abs(min_sum):
            min_sum = curr_sum
            result = (arr[left], arr[right])
        if curr_sum < 0:
            left += 1
        else:
            right -= 1
    return result
print(closest_to_zero([1, 60, -10, 70, -80, 85]))  # Problem 5


# Problem 6: Three elements whose sum is equal to a given number
def find_three_sum(arr, target):
    arr.sort()
    for i in range(len(arr) - 2):
        left, right = i + 1, len(arr) - 1
        while left < right:
            current_sum = arr[i] + arr[left] + arr[right]
            if current_sum == target:
                return (arr[i], arr[left], arr[right])
            elif current_sum < target:
                left += 1
            else:
                right -= 1
    return None
print(find_three_sum([12, 3, 4, 1, 6, 9], 24))  # Problem 6


# Problem 7: Three elements i, j, k such that i^2 + j^2 = k^2
def find_pythagorean_triplet(arr):
    squared_arr = [x ** 2 for x in arr]
    squared_set = set(squared_arr)
    for i in range(len(squared_arr)):
        for j in range(i + 1, len(squared_arr)):
            if squared_arr[i] + squared_arr[j] in squared_set:
                return (arr[i], arr[j], int((squared_arr[i] + squared_arr[j]) ** 0.5))
    return None
print(find_pythagorean_triplet([3, 1, 4, 6, 5]))  # Problem 7

# Problem 8: Majority element (more than n/2 times)
def find_majority_element(arr):
    count = 0
    candidate = None
    for num in arr:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)
    if arr.count(candidate) > len(arr) // 2:
        return candidate
    return None
print(find_majority_element([2, 2, 1, 1, 2, 2, 2]))  # Problem 8

# Problem 9: Row with the maximum number of 0â€™s in an n x n matrix
def row_with_max_zeros(matrix):
    max_zeros = -1
    max_row_index = -1
    for i, row in enumerate(matrix):
        zeros_count = len(row) - sum(row)
        if zeros_count > max_zeros:
            max_zeros = zeros_count
            max_row_index = i
    return max_row_index
print(row_with_max_zeros([[1, 1, 1, 0], [1, 0, 0, 0], [1, 1, 0, 0]]))  # Problem 9


# Problem 10: Sort array of 0's, 1's, and 2's
def sort_012(arr):
    low, mid, high = 0, 0, len(arr) - 1
    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:
            arr[high], arr[mid] = arr[mid], arr[high]
            high -= 1
    return arr

print(sort_012([0, 1, 1, 0, 1, 2, 1, 2, 0, 0, 0, 1]))  # Problem 10
```
